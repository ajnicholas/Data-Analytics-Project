```python
import re
import string
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download("stopwords")
nltk.download('wordnet')
stop_words = set(stopwords.words("english"))
```

```python
def cleantweets(text):
    if isinstance(text, str):
        text = text.lower()
        text_no_links = re.sub(r'http\S+', '', text)
        text_no_at = re.sub(r'@\w+', '', text_no_links)
        text_no_hashtags = re.sub(r'#\w+', '', text_no_at)
        cleaned_text = re.sub(r'[\n\r]', ' ', text_no_hashtags)
        return cleaned_text
    else:
        return text
df["cleaned original tweets"] = df['OriginalTweet'].apply(cleantweets)
df.tail(10)
```

```python
from nltk.tokenize import TweetTokenizer
from nltk.corpus import stopwords
tokenizer = TweetTokenizer()
nltk_stopwords = set(stopwords.words('english'))
def cleaned_tokenize_text(text):
  if isinstance(text, str):
    tokenized_text = tokenizer.tokenize(text)
    cleaned_tokens = [token.lower() for token in tokenized_text if token.lower() not in nltk_stopwords and not re.match(r'[^\w\s]', token)]
    return cleaned_tokens
  else:
    return []
  import string
df["Clean_TokenizedTweets"] = df['cleaned original tweets'].apply(cleaned_tokenize_text)
df.tail(10)
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
def stem_tweets(text):
    if isinstance(text, str):
        tokens = text.split()
        stemmed_tokens = [stemmer.stem(token) for token in tokens]
        stemmed_text = ' '.join(stemmed_tokens)
        return stemmed_text
    else:
        return []
df["stemmedtweets"] = df['cleaned original tweets'].apply(stem_tweets)
df.head(5)
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
df = pd.read_csv('Corona_NLP_test.csv')
df['OriginalTweet'].fillna('', inplace=True)
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['OriginalTweet'])
print(tfidf_matrix)
```

```python
from sklearn.model_selection import train_test_split
X = df['OriginalTweet']
y = df['Sentiment']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
print("Training set size:", len(X_train))
print("Test set size:", len(X_test))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
df['OriginalTweet'].fillna('', inplace=True)
df['Sentiment'].fillna('', inplace=True)
X = df['OriginalTweet']
y = df['Sentiment']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
logistic_regression_model = LogisticRegression(max_iter=1000)
logistic_regression_model.fit(X_train_tfidf, y_train)
y_pred = logistic_regression_model.predict(X_test_tfidf)
accuracy = accuracy_score(y_test, y_pred)
classification_report_str = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy}") #Accuracy is 57.26%
print(f"Classification Report:\n{classification_report_str}")
```

```python
import pandas as pd
import numpy as np
from scipy import stats
df = pd.read_csv('q3_data.csv')
mean_x1 = df['x1'].mean()
mean_x2 = df['x2'].mean()
mode_x1 = stats.mode(df['x1']).mode
mode_x2 = stats.mode(df['x2']).mode
median_x1 = df['x1'].median()
median_x2 = df['x2'].median()
Q1_x1 = df['x1'].quantile(0.25)
Q3_x1 = df['x1'].quantile(0.75)
IQR_x1 = Q3_x1 - Q1_x1
Q1_x2 = df['x2'].quantile(0.25)
Q3_x2 = df['x2'].quantile(0.75)
IQR_x2 = Q3_x2 - Q1_x2
print(f"Mean x1: {mean_x1}")
print(f"Mean x2: {mean_x2}")
print(f"Mode x1: {mode_x1}")
print(f"Mode x2: {mode_x2}")
print(f"Median x1: {median_x1}")
print(f"Median x2: {median_x2}")
print(f"IQR x1: {IQR_x1}")
print(f"IQR x2: {IQR_x2}")
```
